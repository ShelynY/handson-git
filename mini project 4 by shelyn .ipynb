```python
# deteksi outlier dengan IQR 
# Hitung Q1,Q3 dan IQR 
Q1 = df[numerik].quantile(0.25)
Q3 = df[numerik].quantile(0.75)
IQR = Q3 - Q1

# tentukan batas bawah dan atas 
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR 

# hapus baris yang memiliki outlier di salah satu kolom numerik 
df_outlier   = df[((df[numerik] < lower_bound) | (df[numerik] > upper_bound)).any(axis=1)]

print("Data Outlier: ",df_outlier.shape[0]," baris")
```

```python
import pandas as pd
import numpy as np
from datetime import datetime

# FEATURE ENGINEERING SIMPLE + EXPORT UNTUK POWER BI

# 1. BASIC METRICS
df['Attrition'] = df['Resigned'].astype(int)
df['Attrition_Rate'] = df['Resigned'].mean() * 100  # Untuk overall metric

# 2. WORKLOAD & BURNOUT FEATURES
df['Total_Workload_Hours'] = df['Work_Hours_Per_Week'] + df['Overtime_Hours']
df['Work_Intensity'] = df['Total_Workload_Hours'] / 40  # 40 jam normal
df['Productivity_Rate'] = df['Projects_Handled'] / df['Work_Hours_Per_Week']

# Burnout Risk Score (SIMPLE VERSION)
df['Burnout_Risk_Score'] = (
    (df['Overtime_Hours'] / 20) * 0.4 +           # Overtime contribution
    (df['Work_Hours_Per_Week'] / 60) * 0.3 +      # Weekly hours contribution  
    (df['Sick_Days'] / 10) * 0.3                  # Health impact
).clip(0, 1)

# Burnout Risk Category
df['Burnout_Risk_Category'] = pd.cut(
    df['Burnout_Risk_Score'],
    bins=[0, 0.3, 0.7, 1],
    labels=['Low Burnout Risk', 'Medium Burnout Risk', 'High Burnout Risk']
)

# 3. REMOTE WORK CATEGORIZATION
df['Remote_Work_Type'] = np.select(
    [
        df['Remote_Work_Frequency'] == 0,
        df['Remote_Work_Frequency'] == 100,
        df['Remote_Work_Frequency'] <= 50
    ],
    ['Full Office', 'Full Remote', 'Hybrid'],
    default='Hybrid'
)

# 4. SALARY FAIRNESS METRICS
df['Salary_Quartile'] = pd.qcut(df['Monthly_Salary'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
df['Salary_to_Performance_Ratio'] = df['Monthly_Salary'] / df['Performance_Score']

# 5. CAREER PROGRESSION
df['Promotion_Rate'] = df['Promotions'] / (df['Years_At_Company'] + 1)
df['Performance_Trend'] = df['Performance_Score'] / (df['Years_At_Company'] + 1)


# Export semua data ke CSV
output_filename = 'Employee_Analytics_Simple.csv'
df.to_csv(output_filename, index=False)
```
